# Membaca dataset

Fungsi untuk mengubah file xml menjadi dataframe agar lebih mudah diolah.

```python
import pandas as pd 
import xml.etree.ElementTree as et 

def xml_to_df(nama_file_xml, class_exist):
  xtree = et.parse(nama_file_xml)
  xroot = xtree.getroot()

  rows = []
  cols = []
  if class_exist:
    cols = ["id", "class", "judul", "isi"]
  else:
    cols = ["id", "judul", "isi"]

  for dok in xroot: 
      dok_id = dok.find("ID").text
      dok_pertanyaan = dok.find("PERTANYAAN")
      dok_judul = dok_pertanyaan.find("JUDUL").text
      dok_isi = dok_pertanyaan.find("ISI").text

      if class_exist:
        dok_class = dok.find("CLASS")
        lst_temp = []
        for label in dok_class:
          lst_temp.append(label.text)
        dok_class = tuple(lst_temp)
        
        rows.append({"id": dok_id, "class": dok_class, "judul": dok_judul, "isi": dok_isi})
      else:
        rows.append({"id": dok_id, "judul": dok_judul, "isi": dok_isi})

  df = pd.DataFrame(rows, columns = cols)
  return df
```

## Normalisasi Text dan Stopword Removal

```python
import re

# Fungsi untuk melakukan normalisasi teks

def normalize(text):
    res = text.lower().strip() # Mengubah uppercase menjadi lowercase dan melakukan trimming pada teks
    res = re.sub(r'((www\.[^\s]*)|(https : //[^\s]*)|(http : //[^\s]*))', '', res) # Menghapus URL www.xxx atau https : xxx atau http : xxx
    res = re.sub(r'((\S* @ \S*\s?)|(\S*@\S*\s?))', '', res) # Menghapus email address xxx@xxx dan xxx @ xxx
    normal_regex = re.compile(r"(.)\1{1,}") # Regex huruf yang berulang seperti haiiii
    res = normal_regex.sub(r"\1\1", res) # Membuang huruf yang berulang
    res = re.sub("\d", "", res) # Menghilangkan angka
    res = re.sub("[^\w\s]", " ", res) # Menghilangkan tanda baca
    res = re.sub("\s+", " ", res) # Menghilangkan spasi berlebih
    return res
```

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

stop_factory = StopWordRemoverFactory()
lst_stopword = stop_factory.get_stop_words()
own_dict = ["dok", "selamat", "pagi", "siang", "sore", "malam", "dokter", "hehehe"]

# Fungsi untuk menghilangkan stopword dari suatu teks

def remove_stopword(text):
    lst_token = text.split(" ")
    res = []
    for token in lst_token:
        if token not in lst_stopword and token not in own_dict:
            res.append(token)
    return " ".join(res)
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemming process
def stemming(text):
    hasil_stemming = stemmer.stem(text)
    return hasil_stemming
```

```python
# Fungsi normalisasi teks dan menghilangkan stopword (preprocess) dataset

def preprocess_dataset(dataset):
  for index, row in dataset.iterrows():
    judul = row['judul']
    isi = row['isi']
    normalized_text_judul = normalize(judul)
    normalized_text_isi = normalize(isi)
    cleaned_text_judul = remove_stopword(normalized_text_judul)
    cleaned_text_isi = remove_stopword(normalized_text_isi)
    stemmed_text_judul = stemming(cleaned_text_judul)
    stemmed_text_isi = stemming(cleaned_text_isi)
    row['judul'] = stemmed_text_judul
    row['isi'] = stemmed_text_isi
  return dataset
```

```python
# Meng-assign kolom spesialisasi sesuai class untuk setiap instance

def assign_class(df):
  for index, row in df.iterrows():
    label = row['class']
    for i in range(len(label)):
      if (label[i] == "Kebidanan dan Kandungan"):
        df.at[index,'Kebidanan dan Kandungan']= 1
      elif (label[i] == "Penyakit Dalam"):
        df.at[index,'Penyakit Dalam']= 1
      elif (label[i] == "Kesehatan Anak"):
        df.at[index,'Kesehatan Anak']= 1
      elif (label[i] == "Kesehatan Kulit dan Kelamin"):
        df.at[index,'Kesehatan Kulit dan Kelamin']= 1
      elif (label[i] == "Kesehatan Gizi"):
        df.at[index,'Kesehatan Gizi']= 1
      elif (label[i] == "Kesehatan Telinga, Hidung dan Tenggorokan (THT)"):
        df.at[index,'Kesehatan Telinga, Hidung dan Tenggorokan (THT)']= 1
      elif (label[i] == "Gigi"):
        df.at[index,'Gigi']= 1
      elif (label[i] == "Kesehatan Mata"):
        df.at[index,'Kesehatan Mata']= 1
      elif (label[i] == "Bedah"):
        df.at[index,'Bedah']= 1
      elif (label[i] == "Kesehatan Jiwa"):
        df.at[index,'Kesehatan Jiwa']= 1
      elif (label[i] == "Ortopedi (Tulang)"):
        df.at[index,'Ortopedi (Tulang)']= 1
      elif (label[i] == "Jantung dan Pembuluh Darah"):
        df.at[index,'Jantung dan Pembuluh Darah']= 1
      elif (label[i] == "Urologi"):
        df.at[index,'Urologi']= 1
      elif (label[i] == "Saraf"):
        df.at[index,'Saraf']= 1
      elif (label[i] == "Pulmonologi (Paru)"):
        df.at[index,'Pulmonologi (Paru)']= 1
      elif (label[i] == "Umum"):
        df.at[index,'Umum']= 1
  return df
```

### Problem Transformation

Model dibangun menggunakan teknik *Problem Transformation* untuk multi-label classification.

Terdapat tiga teknik pada Problem Transformation, yaitu
* Binary relevance
* Classifier chains
* Label powerset

referensi: https://www.section.io/engineering-education/multi-label-classification-with-scikit-multilearn/

#### Binary relevance
It refers to converting the multi-label dataset into a single-label dataset.

```python
corpus_testing_data = testing_data_df_cleaned["judul"] + " " + testing_data_df_cleaned["isi"]
Xtest = vectorizer.transform(corpus_testing_data).toarray()
```

```python
def testing_data_prediction(model):
  pred_res_testing = model.predict(Xtest)
  pred_res_testing.toarray()

  data = pred_res_testing.toarray()

  df_hasil_pred = testing_data_df_cleaned.drop(["judul", "isi"], axis = 1)

  df_hasil_pred['Kebidanan dan Kandungan'] = data[:, 0]
  df_hasil_pred['Penyakit Dalam'] = data[:, 1]
  df_hasil_pred['Kesehatan Anak'] = data[:, 2]
  df_hasil_pred['Kesehatan Kulit dan Kelamin'] = data[:, 3]
  df_hasil_pred['Kesehatan Gizi'] = data[:, 4]
  df_hasil_pred['Kesehatan Telinga, Hidung dan Tenggorokan (THT)'] = data[:, 5]
  df_hasil_pred['Gigi'] = data[:, 6]
  df_hasil_pred['Kesehatan Mata'] = data[:, 7]
  df_hasil_pred['Bedah'] = data[:, 8]
  df_hasil_pred['Kesehatan Jiwa'] = data[:, 9]
  df_hasil_pred['Ortopedi (Tulang)'] = data[:, 10]
  df_hasil_pred['Jantung dan Pembuluh Darah'] = data[:, 11]
  df_hasil_pred['Urologi'] = data[:, 12]
  df_hasil_pred['Saraf'] = data[:, 13]
  df_hasil_pred['Pulmonologi (Paru)'] = data[:, 14]
  df_hasil_pred['Umum'] = data[:, 15]
  return df_hasil_pred
```
